```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt
```

```python
# Stock ticker
url = 'TSLA'

# Set start and end date for stock price 
start = dt.datetime(2015, 1, 1)
end = dt.datetime.now()

# Creates empty dataframe and retrieves stock data
stock_df = pd.DataFrame()
stock_df = yf.download(url, start, end)
```

```python
# Calculating Exponentially Weighted Moving Averages 

emas = [3,5,8,10,12,15,30,35,40,45,50,60]
for ema in emas:

    # for each rolling window, the moving average is calculated using the closing price 
    stock_df["EMA_{}".format(ema)] = round(stock_df.iloc[:,4].ewm(span=ema, adjust=False).mean(),2)
 
stock_df.tail()
```

```python
pos = 0
num = 0
percent_change = []


for i in stock_df.index:
    cmin = min(stock_df["EMA_3"][i],stock_df["EMA_5"][i],stock_df["EMA_8"][i],stock_df["EMA_10"][i],stock_df["EMA_12"][i],stock_df["EMA_15"][i])
    cmax = max(stock_df["EMA_30"][i],stock_df["EMA_35"][i],stock_df["EMA_40"][i],stock_df["EMA_45"][i],stock_df["EMA_50"][i],stock_df["EMA_60"][i])

    close = stock_df["Adj Close"][i]

    if (cmin > cmax):
        print("Red White Blue")
        if(pos == 0):
            bp = close
            pos = 1
            print("Buying now at " + str(bp))

    elif (cmin < cmax):
        print("Blue White Red")
        if(pos == 1):
            pos = 0
            sp = close
            print("Selling now at " + str(sp))
            pc = (sp/bp-1)*100
            percent_change.append(pc)

if(num == stock_df["Adj Close"].count() - 1 and pos == 1):
    pos = 0
    sp = close
    print("Selling now at " + str(sp))
    pc = (sp/bp -1) * 100
    percent_change.append(pc)

print(percent_change)
```

```python
# Print Summary Statistics

gains = 0
number_gains = 0
losses = 0
number_losses = 0
total_return = 1

for i in percent_change:
    if (i > 0):
        gains += i
        number_gains += 1
    else:
        losses += i
        number_losses += 1
    total_return = total_return * ((i/100 + 1))

total_return = round((total_return - 1) * 100,2)

if (number_gains > 0):
    average_gain = gains/number_gains
    max_return = str(max(percent_change))
else:
    average_gain = 0
    max_return = "undefined"

if (number_losses > 0):
    average_loss = losses/number_losses
    max_loss = str(min(percent_change))
    ratio = str(-average_gain/average_loss)
else:
    average_loss = 0
    max_loss = "undefined"
    ratio = "infinite"

if (number_gains > 0 or number_losses > 0):
    batting_average = number_gains/(number_gains + number_losses)
else: 
    batting_average = 0

print("Results for " + url + " going back to " + str(stock_df.index[0]) + ", Sample size: " + str(number_gains + number_losses) + "")
print("EMA used: " + str(emas))
print("Batting Average: " + str(batting_average))
print("Gain/loss ratio " + ratio)
print("Average Gain: " + str(average_gain))
print("Average Loss: "+ str(average_loss))
print("Max Return " + str(max_return))
print("Total return over " + str(number_gains + number_losses) + " trades: " + str(total_return) + "%")
```
